```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
with open('../data/json_sp500.json', 'r') as f:
    datos = json.load(f)

try:
    accion = input("Introduce tiker de accion:")
    datos_accion = datos[accion]
    tipo = input("Elige el dato a imprimir(Close,High,Low,Open,Volume):")
    x=[]
    y=[]

    for dia in datos_accion:
        x.append(dia["('Date', '')"])
        y.append(dia[f"('{tipo}', '{accion}')"])   
    plt.plot(x, y, color='blue', linewidth=2)
    plt.title(f'Gráfica accion sp500 ({accion})')
    plt.xlabel('Fecha')
    plt.ylabel(f'Valor_sp500({tipo})')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    ax = plt.gca()
    ax.set_xticks(ax.get_xticks()[::12]) 
    plt.show()
except Exception as e:
    print(f"Error en {accion}: {e}")



```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
ticker = "^GSPC"

data = yf.download(ticker, period="1y")
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Cierre S&P 500', color='royalblue', linewidth=2)
plt.title('Evolución del S&P 500 - Último Año', fontsize=14)
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
